Below we plot the average episode length as DQN trains on finding the shortest path between two nodes in a random graph. The graphs are Barabasi-Albert graphs, parameterized by $n, m_0, m$. $n$ is the number of nodes in the graph, $m_0$ is the number of initial nodes, and $m$ is approximately the lower bound of the average number of neighbors of a node. These parameters are [further described on wikipedia](https://en.wikipedia.org/wiki/Barab%C3%A1si%E2%80%93Albert_model#Algorithm)


The experiments were run with the following parameters:

| Experiment name | $n$ | $m_0$ | $m$ |
| ----------- | ----------- | ----------- | ----------- |
| Experiment 4      | 30       | 8      | 6       |
| Experiment 5      | 60       | 20      | 16       |
| Experiment 6      | 60       | 16      | 15       |
| Experiment 7      | 40       | 6      | 6       |


```python
#import gym
#import gym_graph_search

import numpy as np
import csv

import matplotlib.pyplot as plt

plt.figure(1, figsize=(12,10)) 

jobids = list(range(4,8))
filenames = ["experiment_logs/experiment{}/progress.csv".format(jobid) for jobid in jobids]

for _, filename in enumerate(filenames):
    read_in = list(csv.reader(open(filename, "r"), delimiter=","))[1:]
    #print(read_in)
    exp_data = np.array(read_in).astype("float")

    # Entries in CSV are as follows:
    # exploration epsilon, episode #, avg reward over past <=100 episodes, total steps taken
    mean_ep_len = [-ls[2] for ls in exp_data]
    
    ep_num = [ls[1] for ls in exp_data]
    step_num = [ls[3] for ls in exp_data]

    plt.subplot(211)
    plt.plot(ep_num, mean_ep_len, label = filename)
    plt.subplot(212)
    plt.plot(step_num, mean_ep_len, label = filename)

    
plt.subplot(211)
plt.xlabel("Number of training episodes completed")
plt.ylabel("Average episode length")
plt.legend()
plt.subplot(212)
plt.xlabel("Number of training steps taken")
plt.ylabel("Average episode length")
plt.legend()
plt.show()

```
